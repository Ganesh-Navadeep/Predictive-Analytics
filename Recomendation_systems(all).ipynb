```python
#1.Product Recommendation System

def jaccard_similarity(set1, set2):
    intersection = set1.intersection(set2)
    union = set1.union(set2)
    return len(intersection) / len(union) if union else 0

# Category keyword mapping
product_interest_map = {
    "saucepan": "cooking utensils",
    "knife": "cooking utensils",
    "blender": "cooking utensils",
    "diapers": "baby ware",
    "baby bottle": "baby ware",
    "vitamins": "medicine",
    "painkiller": "medicine",
    "jeans": "clothing",
    "shirt": "clothing",
    "dress": "clothing",
    "lipstick": "styling",
    "foundation": "styling",
    "t-shirt": "garments",
    "jacket": "garments",
    "bra": "inners",
    "boxers": "inners",
    "phone": "electronics",
    "laptop": "electronics",
    "mouse": "electronics",
    "yarn": "crafts",
    "paintbrush": "crafts",
    "chips": "food",
    "pasta": "food"
}

def predict_interests(shopping_list):
    interests = set()
    for item in shopping_list:
        category = product_interest_map.get(item)
        if category:
            interests.add(category)
    return interests

def get_user_input():
    username = input("Enter your name: ")
    shopping_input = input("Enter your shopping history (comma-separated): ")
    shopping_list = set(map(str.strip, shopping_input.lower().split(',')))
    interests = predict_interests(shopping_list)
    return username, shopping_list, interests

def recommend_products(current_user, users_data):
    best_match_score = 0
    best_product = None

    for user, data in users_data.items():
        if user == current_user[0]:
            continue
        sim_score = jaccard_similarity(current_user[1], data["shopping_history"])
        if sim_score > 0:
            new_products = data["shopping_history"] - current_user[1]
            for product in new_products:
                if sim_score > best_match_score:
                    best_match_score = sim_score
                    best_product = product

    # Fallback: pick any one unseen product from the whole dataset
    if not best_product:
        all_products = set()
        for data in users_data.values():
            all_products.update(data["shopping_history"])
        unseen = all_products - current_user[1]
        best_product = next(iter(unseen)) if unseen else "generic item"

    return best_product


# Large, diverse dataset (no explicit interests)
users_data = {
    "alice": {
        "shopping_history": {"saucepan", "knife", "blender", "chips", "pasta"}
    },
    "bob": {
        "shopping_history": {"diapers", "baby bottle", "pasta", "yarn"}
    },
    "carol": {
        "shopping_history": {"jeans", "shirt", "lipstick", "foundation", "pasta"}
    },
    "dave": {
        "shopping_history": {"bra", "boxers", "jacket", "t-shirt"}
    },
    "emma": {
        "shopping_history": {"phone", "laptop", "mouse"}
    },
    "frank": {
        "shopping_history": {"painkiller", "vitamins", "chips"}
    },
    "grace": {
        "shopping_history": {"paintbrush", "yarn", "laptop"}
    },
    "hannah": {
        "shopping_history": {"dress", "lipstick", "jacket"}
    },
    "ivan": {
        "shopping_history": {"shirt", "mouse", "phone", "jeans"}
    },
    "julia": {
        "shopping_history": {"diapers", "baby bottle", "jeans"}
    }
}

# Predict interests for existing users
for user, data in users_data.items():
    data["interests"] = predict_interests(data["shopping_history"])

if __name__ == "__main__":
    username, shopping_list, interests = get_user_input()
    users_data[username] = {
        "shopping_history": shopping_list,
        "interests": interests
    }

    recommendations = recommend_products((username, shopping_list), users_data)
    predicted_interests = predict_interests(shopping_list)
    print(f"\nPredicted interests for {username.title()}:")
    for interest in interests:
      print(f"- {interest.title()}")

    print(f"\nRecommended product for {username.title()}:")
    print(f"- {recommendation.title()}")
```

```python
#3. Student Course Suggestion System

def jaccard_similarity(set1, set2):
    intersection = set1.intersection(set2)
    union = set1.union(set2)
    return len(intersection) / len(union) if union else 0

# Course-to-interest mapping
course_interest_map = {
    "python basics": {"programming", "foundations"},
    "data analysis": {"data science", "statistics"},
    "intro to ai": {"ai", "machine learning"},
    "neural networks": {"ai", "deep learning"},
    "marketing fundamentals": {"business", "marketing"},
    "consumer psychology": {"psychology", "marketing"},
    "philosophy 101": {"philosophy"},
    "design thinking": {"design", "innovation"},
    "graphic design": {"design", "creative"},
    "web development": {"programming", "web"},
    "frontend frameworks": {"programming", "web"},
    "cloud essentials": {"cloud", "infrastructure"},
    "cybersecurity basics": {"cybersecurity", "it"},
    "ui ux foundations": {"design", "ux"},
    "public speaking": {"communication"},
    "nutrition science": {"health"},
    "creative writing": {"writing", "literature"},
    "machine learning": {"ai", "data science"},
    "time management": {"personal development"},
    "entrepreneurship": {"business", "startup"}
}

# Simulated student database
students_data = {
    "alice": {
        "courses_taken": {"python basics", "web development", "frontend frameworks"},
        "focus_area": "web"
    },
    "bob": {
        "courses_taken": {"data analysis", "machine learning", "intro to ai"},
        "focus_area": "ai"
    },
    "carol": {
        "courses_taken": {"marketing fundamentals", "consumer psychology"},
        "focus_area": "marketing"
    },
    "dave": {
        "courses_taken": {"graphic design", "ui ux foundations", "design thinking"},
        "focus_area": "design"
    },
    "emma": {
        "courses_taken": {"philosophy 101", "creative writing"},
        "focus_area": "philosophy"
    },
    "frank": {
        "courses_taken": {"cloud essentials", "cybersecurity basics"},
        "focus_area": "it"
    },
    "grace": {
        "courses_taken": {"nutrition science", "time management"},
        "focus_area": "health"
    },
    "hannah": {
        "courses_taken": {"entrepreneurship", "marketing fundamentals", "public speaking"},
        "focus_area": "business"
    }
}

def detect_interests(courses_taken):
    detected = set()
    for course in courses_taken:
        categories = course_interest_map.get(course)
        if categories:
            detected.update(categories)
    return detected

def get_user_input():
    username = input("Enter your name: ")
    courses_input = input("Enter courses you've completed (comma-separated): ")
    focus_area = input("Enter your current focus area (e.g., ai, business, design): ").lower()

    courses_taken = set(map(str.strip, courses_input.lower().split(',')))
    interests = detect_interests(courses_taken)

    return username, courses_taken, interests, focus_area

def recommend_courses(current_student, students_data):
    username, student_courses, student_interests, student_focus = current_student

    scored_courses = {}

    for student, data in students_data.items():
        if student == username:
            continue

        sim_score = jaccard_similarity(student_courses, data["courses_taken"])
        if sim_score == 0:
            continue

        other_interests = detect_interests(data["courses_taken"])
        if not student_interests.intersection(other_interests):
            continue
        if student_focus != data["focus_area"]:
            continue

        new_courses = data["courses_taken"] - student_courses
        for course in new_courses:
            scored_courses[course] = max(sim_score, scored_courses.get(course, 0))

    # Fallback if no match found
    if not scored_courses:
        all_courses = set(course_interest_map.keys())
        fallback = [
            c for c in all_courses - student_courses
            if student_focus in course_interest_map.get(c, set())
        ]
        return fallback[:4] if fallback else list(all_courses - student_courses)[:4]

    # Sort by similarity score and return top 2–4
    sorted_courses = sorted(scored_courses.items(), key=lambda x: x[1], reverse=True)
    recommended = [course for course, _ in sorted_courses][:4]

    return recommended


# Main program
if __name__ == "__main__":
    username, courses_taken, interests_detected, focus = get_user_input()
    students_data[username] = {
        "courses_taken": courses_taken,
        "focus_area": focus
    }

    print(f"\nDetected interests for {username.title()}:")
    for interest in interests_detected:
        print(f"- {interest.title()}")

    recommendations = recommend_courses((username, courses_taken, interests_detected, focus), students_data)

    print(f"\nRecommended courses for {username.title()}:")
    for course in recommendations[:4]:
      print(f"- {course.title()}")
```

```python
#5. health risk prediction model

def get_user_input():
    print("Please answer the following questions honestly.\n")

    stress_level = input("Stress level (low/medium/high): ").lower()
    sleep_hours = float(input("Average sleep hours per night (e.g., 7.5): "))

    print("\nLifestyle Indicators (yes/no):")
    poor_diet = input("Do you have an unhealthy or irregular diet? ").lower() == "yes"
    sedentary = input("Is your lifestyle mostly sedentary (low physical activity)? ").lower() == "yes"
    high_screen_time = input("Do you use screens (phone/laptop) more than 6 hrs/day? ").lower() == "yes"
    socially_isolated = input("Do you have little to no social interaction daily? ").lower() == "yes"

    return {
        "stress": stress_level,
        "sleep": sleep_hours,
        "diet": poor_diet,
        "sedentary": sedentary,
        "screen_time": high_screen_time,
        "isolation": socially_isolated
    }

def assess_mental_health_risk(data):
    score = 0

    # Stress level
    if data["stress"] == "high":
        score += 3
    elif data["stress"] == "medium":
        score += 2
    else:
        score += 1

    # Sleep
    if data["sleep"] < 6 or data["sleep"] > 10:
        score += 2
    elif 6 <= data["sleep"] <= 8:
        score += 0  # healthy range
    else:
        score += 1

    # Lifestyle
    lifestyle_flags = sum([
        data["diet"],
        data["sedentary"],
        data["screen_time"],
        data["isolation"]
    ])
    score += lifestyle_flags * 1.5

    # Risk classification
    if score <= 4:
        risk = "Low"
        note = "Your inputs suggest good mental health habits."
    elif 5 <= score <= 7.5:
        risk = "Moderate"
        note = "You show some risk factors. Consider improving sleep and lifestyle balance."
    else:
        risk = "High"
        note = "Multiple risk indicators detected. Mental health support or lifestyle changes may help."

    return risk, score, note

# Run the program
if __name__ == "__main__":
    user_data = get_user_input()
    risk_level, score, explanation = assess_mental_health_risk(user_data)

    print("\n🧠 Mental Health Risk Assessment")
    print(f"Risk Level: {risk_level}")
    print(f"Score: {score}/10+")
    print(f"Note: {explanation}")
```

```python
#6.  Disease Likelihood Prediction System based on clinical parameters.
def get_patient_data():
    print("📋 Patient Clinical Data Input\n")

    age = int(input("Age: "))
    systolic_bp = int(input("Systolic Blood Pressure (mmHg): "))
    cholesterol = int(input("Total Cholesterol (mg/dL): "))
    glucose = int(input("Fasting Glucose Level (mg/dL): "))
    bmi = float(input("Body Mass Index (BMI): "))

    print("\nLifestyle Indicators (yes/no):")
    family_history = input("Family history of similar diseases? ").lower() == "yes"
    smoker = input("Smoker? ").lower() == "yes"
    alcohol = input("Consumes alcohol frequently? ").lower() == "yes"
    sedentary = input("Low physical activity? ").lower() == "yes"

    return {
        "age": age,
        "bp": systolic_bp,
        "cholesterol": cholesterol,
        "glucose": glucose,
        "bmi": bmi,
        "family_history": family_history,
        "smoker": smoker,
        "alcohol": alcohol,
        "sedentary": sedentary
    }

def predict_disease_risk(data):
    risk_score = 0

    # Age
    if data["age"] >= 60:
        risk_score += 2
    elif data["age"] >= 45:
        risk_score += 1

    # Blood pressure
    if data["bp"] >= 140:
        risk_score += 2
    elif data["bp"] >= 130:
        risk_score += 1

    # Cholesterol
    if data["cholesterol"] >= 240:
        risk_score += 2
    elif data["cholesterol"] >= 200:
        risk_score += 1

    # Glucose
    if data["glucose"] >= 126:
        risk_score += 2
    elif data["glucose"] >= 100:
        risk_score += 1

    # BMI
    if data["bmi"] >= 30:
        risk_score += 2
    elif data["bmi"] >= 25:
        risk_score += 1

    # Lifestyle & genetics
    if data["family_history"]:
        risk_score += 2
    if data["smoker"]:
        risk_score += 1.5
    if data["alcohol"]:
        risk_score += 1
    if data["sedentary"]:
        risk_score += 1.5

    # Classification
    if risk_score <= 4:
        risk_level = "Low Risk"
        message = "👍 Based on current parameters, your disease risk is low. Keep up healthy habits!"
    elif risk_score <= 7.5:
        risk_level = "Moderate Risk"
        message = "⚠️ Some risk factors are present. Consider lifestyle changes and regular check-ups."
    else:
        risk_level = "High Risk"
        message = "🚨 Multiple clinical risk factors detected. Medical evaluation is strongly advised."

    return risk_level, round(risk_score, 1), message

# Run the tool
if __name__ == "__main__":
    patient_data = get_patient_data()
    risk_level, score, msg = predict_disease_risk(patient_data)

    print("\n🩺 Disease Risk Prediction Result")
    print("-" * 40)
    print(f"Risk Category : {risk_level}")
    print(f"Risk Score    : {score}/10+")
    print(f"Insight       : {msg}")
    print("-" * 40)
```

```python
#7. Demand Forecasting Model for Warehouse Inventory Planning.

def get_sales_data():
    print("📦 Warehouse Product Demand Input")
    print("-" * 40)

    products = {}

    num_products = int(input("How many products to forecast? "))

    for _ in range(num_products):
        name = input("\n🔹 Enter product name: ").strip().lower()
        category = input("   Category (perishable/essential/electronic/apparel/misc): ").strip().lower()
        sales = input("   Last 6 months of sales (comma-separated): ")
        sales = list(map(int, sales.strip().split(',')))

        if len(sales) != 6:
            print("   ⚠️ Must provide 6 monthly values. Skipping.")
            continue

        products[name] = {
            "category": category,
            "sales": sales
        }

    return products

def forecast_demand(products):
    print("\n📈 Forecasting Demand...\n")
    forecast = {}

    for product, data in products.items():
        last_6_months = data["sales"]
        moving_avg = sum(last_6_months[-3:]) // 3  # last 3 months average

        buffer = 0
        if data["category"] == "perishable":
            buffer = 0.1  # low buffer
        elif data["category"] in ["essential", "electronic"]:
            buffer = 0.2
        else:
            buffer = 0.15

        forecast_qty = int(moving_avg * (1 + buffer))

        recommendation = ""
        if forecast_qty > max(last_6_months):
            recommendation = "⬆️ Increase stock"
        elif forecast_qty < min(last_6_months):
            recommendation = "⬇️ Reduce stock"
        else:
            recommendation = "✅ Maintain current levels"

        forecast[product] = {
            "predicted_demand": forecast_qty,
            "trend": recommendation
        }

    return forecast

# Run program
if __name__ == "__main__":
    warehouse_data = get_sales_data()
    prediction = forecast_demand(warehouse_data)

    print("\n📦 Product Demand Forecast Summary")
    print("-" * 40)
    for product, result in prediction.items():
        print(f"🛒 Product        : {product.title()}")
        print(f"📊 Forecast Qty   : {result['predicted_demand']} units")
        print(f"📌 Suggestion     : {result['trend']}")
        print("-" * 40)
```
